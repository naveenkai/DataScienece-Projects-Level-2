```python
import pandas as pd
import numpy as no
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
## distibution of data

data['Class'].value_counts()
```

```python
legit = data[data.Class==0]
fraud = data[data.Class==1]
```

```python
legit.shape, fraud.shape
```

```python
## statistical measures of data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
## compare the values for both transactions

data.groupby('Class').mean()
```

```python
## under-sampling

## no. of fraudlent transactions --> 492

legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns = ['Class'], axis = 1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
## Model Training

#### Logistic regression ###

model = LogisticRegression()
```

```python
## training Logistic regrssion model with training data

model.fit(xtrain, ytrain)
```

```python
## Model evaluation

xtrain_pred = model.predict(xtrain)
training_acc = accuracy_score(xtrain_pred, ytrain)
```

```python
training_acc
```

```python
xtest_pred = model.predict(xtest)
test_acc = accuracy_score(xtest_pred, ytest)
```
